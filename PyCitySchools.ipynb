```python
import os
import pandas as pd
import numpy as np
```

```python
school_df = pd.read_csv(r"C:Resources\schools_complete.csv")
school_df.head(5)
```

```python
student_df = pd.read_csv(r"C:Resources\students_complete.csv")
student_df.head(5)
```

```python
# Combine the data into a single dataset.  
comp_school_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
comp_school_df.head()
```

```python
school_count = comp_school_df['school_name'].nunique()
school_count
```

```python
student_count = comp_school_df['student_name'].count()
student_count
```

```python
total_budget = comp_school_df['budget'].sum()
total_budget
```

```python
average_math_score = comp_school_df['math_score'].mean().round(2)
average_math_score
```

```python
average_reading_score = comp_school_df['reading_score'].mean().round(2)
average_reading_score
```

```python
passing_math_count = comp_school_df[(comp_school_df["math_score"] >= 70)].count()["student_name"]
passing_math_rate = passing_math_count / float(student_count) * 100
passing_math_rate.round(2)
```

```python
passing_reading_count = comp_school_df[(comp_school_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_rate = passing_reading_count / float(student_count) * 100
passing_reading_rate.round(2)
```

```python
passing_math_reading_count = comp_school_df[
    (comp_school_df["math_score"] >= 70) & (comp_school_df["reading_score"] >= 70)
].count()["student_name"]
total_passing_rate = passing_math_reading_count /  float(student_count) * 100
total_passing_rate.round(2)
```

```python
index = pd.Index(['school_count','student_count','total_budget','average_math_score','average_reading_score','passing_math_rate',
                              'passing_reading_rate','total_passing_rate'])
values = [school_count,student_count,total_budget,average_math_score,average_reading_score,passing_math_rate,
                              passing_reading_rate,total_passing_rate]

district_summary = pd.DataFrame(
    {'Metric': index,
     'Value': values,
    })
# Formatting
district_summary["Value"] = district_summary["Value"].map("{:,}".format)
# district_summary["total_budget"] = district_summary["total_budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
school_types = school_df[['school_name','type']]
school_types
```

```python
student_per_school_counts = school_df[['school_name', 'size']]
student_per_school_counts = student_per_school_counts.rename(columns={"size":"student_count"})
student_per_school_counts
```

```python
per_school_budget = school_df[['school_name', 'budget']] 
per_school_budget = per_school_budget.rename(columns={"budget":"total_budget"})
```

```python
school_df['per_student_budget'] = school_df.apply(lambda x:x['budget']/x['size'], axis=1)
per_school_capita = school_df[['school_name','per_student_budget']]
```

```python
per_school_math_average = comp_school_df[['school_name','math_score']].groupby("school_name").mean()
per_school_math_average = per_school_math_average.rename(columns={"math_score":"average_math_score"}).reset_index()
```

```python
per_school_reading_average = comp_school_df[['school_name','reading_score']].groupby("school_name").mean()
per_school_reading_average = per_school_reading_average.rename(columns={"reading_score":"average_reading_score"}).reset_index()
per_school_reading_average
```

```python
# Calculate the number of students per school with math scores of 70 or higher
# comp_school_df['pass']  = np.where(
#     comp_school_df['math_score'] <= 69, 0, np.where(
#     comp_school_df['math_score'] >  69, 1, -1)) 
students_passing_math = comp_school_df[(comp_school_df["math_score"] >= 70)]
school_students_passing_math = students_passing_math.groupby(["school_name"]).size()
school_students_passing_math
```

```python
# # Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = comp_school_df[(comp_school_df["reading_score"] >= 70)]
school_students_passing_reading = students_passing_reading.groupby(["school_name"]).size()
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = comp_school_df[(comp_school_df["reading_score"] >= 70) & (comp_school_df["math_score"] >= 70)]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
students_passing_both = pd.DataFrame(school_students_passing_math_and_reading, columns=['students_passing_both'])
students_passing_both
```
